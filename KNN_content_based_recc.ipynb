```python
import pandas as pd
import numpy as np
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
cars = pd.read_csv('mtcars.csv')
```

```python
cars.head()
```

```python
cars.columns= ['car_names','milage','cyl','disp','hp','drat','weight','qsec','vs','am','gears','carb']
```

```python
cars
```

```python
cars.head()
```

```python
t = [15,300,160,3.2]
```

```python
X = cars.iloc[:,[1,3,4,6]].values
```

```python
X[:5]
```

```python
nbrs = NearestNeighbors(n_neighbors = 5).fit(X)
```

```python
nbrs.kneighbors([t])[1]
```
